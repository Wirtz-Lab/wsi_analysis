```python
import glob, os
```

```python
src = r'\\fatherserverdw\Q\research\images\clue_batch2\good'
imlist = glob.glob(os.path.join(src,'*.ndpi'))
imlist
```

```python
import pandas as pd
df = pd.DataFrame(imlist,columns=['path'])
df
```

```python
df['filename'] = df['path'].apply(lambda x: os.path.basename(x))
df
```

```python
#dataframe to excel chart
dst = r'\\fatherserverdw\Q\research\images\clue_batch2'
df.to_csv(os.path.join(dst,'rename-good.csv'))
```

```python
from openslide import OpenSlide
```

```python
for ndpi in df['path'].iloc:
    ndpi_obj = OpenSlide(ndpi)
    ndpi_obj.associated_images['macro'].save(ndpi.replace('ndpi','png'))
```

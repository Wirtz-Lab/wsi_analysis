```python
def find_file_paths_with_numbers(numbers, file_paths):
    matching_file_paths = []
    for number in numbers:
        for file_path in file_paths:
            if str(number) in file_path[-8:-4] or str(number) in file_path[-9:-4]:
                matching_file_paths.append(file_path)
                break  # Move to the next number once a match is found
    return matching_file_paths
```

```python
converted_paths = find_file_paths_with_numbers(need_filepath, all_paths)
```

```python
final_he_paths = converted_paths + dont_need_filepath
final_he_paths = natsorted(final_he_paths)
```

```python
def copy_files_to_destination(file_paths, destination_directory):
    for file_path in file_paths:
        if os.path.isfile(file_path):  # Check if it's a file
            shutil.copy(file_path, destination_directory)
```

```python
# Now create a bad dataset by only doing affine translations of the HE images, keep US images the same!
def remove_black_region(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if contours:
        x, y, w, h = cv2.boundingRect(contours[0])
        image = image[y:y + h, x:x + w]
    return image

def create_bad_image_pair(dir1, save_dir1):
    random.seed(42)
    os.makedirs(save_dir1, exist_ok=True)
    files1 = [x for x in os.listdir(dir1) if x.endswith(".png")]
    for i in range(len(files1)):
        file1 = files1[i]
        image1 = cv2.imread(os.path.join(dir1, file1))
        dx = random.randint(-50, -10) if random.random() < 0.5 else random.randint(10, 50) # -50~-10 random vertical or horizontal translations
        dy = random.randint(-50, -10) if random.random() < 0.5 else random.randint(10, 50)
        rows, cols, _ = image1.shape
        M = np.float32([[1, 0, dx], [0, 1, dy]])
        distorted_image1 = cv2.warpAffine(image1, M, (cols, rows))
        distorted_image2 = remove_black_region(distorted_image1)
        distorted_image3 = cv2.resize(distorted_image2,(1024,1024),interpolation=2) #INTER_CUBIC = 2
        save_path1 = os.path.join(save_dir1, file1)
        cv2.imwrite(save_path1, distorted_image3)
    print("Complete!")
```

```python
# Set the directories and destination directories
dir1 = r"\\shelter\Kyu\unstain2stain\tiles\registered_tiles\100_good_images\HE"
save_dir1 = r"\\shelter\Kyu\unstain2stain\tiles\registered_tiles\100_good_images_bad\HE"
create_bad_image_pair(dir1, save_dir1)
```
